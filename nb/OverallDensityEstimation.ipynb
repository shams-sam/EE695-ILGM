# imports

```python
import sys
sys.path.append('../src/')

from collections import OrderedDict
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import torch

from data import get_loader
from models import MLP
from utils import plot_density, to_numpy
```

# config

```python
batch_size = 256
dataset = 'mnist'
epochs = 10
hidden_size = [1024, 2048, 256]
learning_rate = 1e-1
log_intv = 100

device = torch.device('cuda:0')
```

# dataloader and model

```python
trainloader = get_loader(dataset, batch_size, train=True)
testloader = get_loader(dataset, batch_size, train=False)

model = MLP(hidden_size=hidden_size).to(device)
init_model(model)
```

```python
plot_density(model)
```

# training

```python
optim = torch.optim.SGD(model.parameters(), lr=learning_rate)
loss_fn = torch.nn.CrossEntropyLoss().to(device)
```

```python
for epoch in range(epochs):
    
    running_loss = 0.0
    running_acc = 0.0
    running_count = 0
    for i, (Y, X) in enumerate(trainloader):
        Y, X = Y.to(device), X.to(device)
        
        X_ = model(Y)
        loss = loss_fn(X_, X)
        running_loss += loss.item()
        optim.zero_grad()
        loss.backward()
        optim.step()
        pred = X_.argmax(axis=1, keepdim=True)
        running_acc += pred.eq(X.view_as(pred)).sum().item()/len(X)
        running_count += 1
        if i % log_intv == 0:
            print('+'* 80)
            print('[epoch: {:3d}, batch: {:4d}] loss: {:.6f} \t acc: {:.2f}'.format(
                epoch, i, running_loss/running_count, running_acc/running_count))
            print('+'* 80)
            running_loss = 0.0
            running_acc = 0.0
            running_count = 0 
            plot_density(model, bins=20)
```

# validate training

```python
def validate(model, loader, device):
    if not device:
        device = torch.device('cpu')
    
    correct = 0.0
    total = 0
    for Y, X in loader:
        Y, X = Y.to(device), X.to(device)
        X_ = model(Y).argmax(axis=1, keepdim=True)
        correct += X_.eq(X.view_as(X_)).sum().item()
        total += len(X)
        
    return correct/total
```

```python
validate(model, testloader, device)
```

# direct density estimation

```python
def estimate_density(array):
    size = array.shape
    array = array.flatten()
    mean = np.mean(array)
    std = np.std(array)
    return mean, std, size

def to_np(tensor):
    return tensor.detach().cpu().numpy()

def draw_weights(density_estimates):
    weights = OrderedDict()
    for name, (mean, std, size) in density_estimates.items():
        weights[name] = torch.normal(mean, std, size=size)
    return weights
```

```python
model_density_estimates = OrderedDict()
for name, param in model.named_parameters():
    model_density_estimates[name] = estimate_density(to_np(param))
```

```python
drawn_weights = draw_weights(model_density_estimates)
drawn_model = MLP(hidden_size=hidden_size).to(device)
drawn_model.load_state_dict(drawn_weights)
```
